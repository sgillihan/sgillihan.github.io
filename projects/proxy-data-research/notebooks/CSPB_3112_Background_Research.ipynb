### Algorithmic Bias Explained: How Automated Decision-Making Becomes Automated Discrimination

Definition of Algorithmic Bias - Algorithmic bias occurs when automated systems produce unfair outcomes that unjustifiably privilege some groups over others. Even when protected traits are not explicity used, algorithms can rely on proxy variables that reflect structural inequality.

#### Sources of bias:
- Choosing the outcome- designers must define what an algorithm optimizes, and these definitions embed value judgements.
- Choosing input variables- some inputs may act as proxies for protected characteristics
- Biased or non-representative training data- if data reflects past discrimination or underrepresents certain groups, the algorithm will replicate those distortions (see https://datanutrition.org)

#### The paper provides a variety of sector-specific examples of algorithmic bias:
- Healthcare: A hospital system used a cost-prediction algorithm as a proxy for health needs, resulting in Black patients needing to be sicker than White patients to receive additional care.
- Employment: Amazon’s recruiting algorithm downgraded resumes containing the word “women’s” because it was trained on historically male-dominated hiring data.
- Government Programs: Michigan’s unemployment fraud algorithm falsely accused over 40,000 people of fraud, with little meaningful appeal process 
- Education: The UK’s 2020 grading algorithm lowered grades disproportionately for lower-income students by relying heavily on school-level historical performance 
- Credit and Finance: Studies show online mortgage lenders continue to overcharge Black and Brown borrowers compared to similarly situated White borrowers
- Housing: Rental screening algorithms disproportionately exclude applicants from over-policed communities 
- Price Optimization: Dynamic pricing algorithms have charged Asian-American neighborhoods higher prices for tutoring and have produced higher insurance premiums in predominantly Black ZIP codes 

#### Proposed Solutions
- Algorithmic Transparency and Accountability: Require impact assessments and independent data audits. Increase regulatory oversight and technical capacity within government
- Race-Conscious Algorithms: Rather than avoid race, developers should measure and correct for disparities across racial groups to prevent proxy discrimination.
- Algorithmic Greenlining: Use algorithms to actively promote equity.

### Constructing Applicants from Loan-Level Data: A Case Study of Mortgage Applications

This study develops a clustering-based algorithm to identify "cross-applicants", defined as individuals submitting multiple mortgage applications, in anonymized loan-level datasets such as the Home Mortgage Disclosure Act (HMDA) data. ​The algorithm uses hierarchical agglomerative clustering with a nearest neighbor chaining method and optimizes tuning parameters without labeled training data. Applied to HMDA data (2018–2023), the method achieved 92.3% precision, identifying over 314,000 cross-applicant clusters.​ 

Potential Applications:
1. Measuring Fairness: Cross-applicants can help assess fairness in mortgage lending decisions across demographic groups. 
2. Monitoring Credit Conditions: Identifying marginal borrowers can provide insights into lenders' strictness and credit conditions over time and across regions. 
3. Studying Mortgage Shopping Behavior: The dataset can be used to analyze how borrowers compare lenders and how shopping behavior varies across demographics. 

### The Great Scrape: The Clash Between Sraping and Privacy

Scraping is defined as the automated extraction of large amounts of data from the internet, often using bots or web crawlers. It is widely used for AI training, web searching, archiving, and research but raises significant privacy concerns. Core privacy principles are defined as fairness, individual rights, transparency, consent, purpose specification, data minimization, onward transfer, and data security. The article purports that scraping and privacy principles are fundamentally at odds in the current landscape. A complete ban on scraping is impractical, but unregulated scraping poses significant privacy risks. A regulatory framework based on public interest and privacy principles is essential to balance the benefits of scraping with the protection of individual privacy. The article discusses the legalities of scraping in depth, including current European and USA legal case studies and existing laws. The article also touches on the possibility of discrimination that arises from data scraping. Ultimately, four principles are suggested to guide the future of data scraping:

1) Reasonable Risk of Harm Principle: The collection, use, or transfer of scraped personal data should not cause unreasonable risk of harm to individuals, disadvantaged groups, or society.
2) Proportional Benefits Principle: The collection, use, or transfer of scraped personal data should provide meaningful benefits to individuals, disadvantaged groups, and society sufficient to outweigh any risks and proportional to or in excess of the benefits to the scraper.
3) Process Principle: The process for deciding the legitimate uses of scraped personal data should be fair, open, accountable, representative, equitable, and deliberative.
4) Protections Principle: Scraped data should be afforded all the same protections as other personal data under privacy laws
unless particular protections impose an unreasonable conflict with other public interest goals.

### A Common Pool of Privacy Problems: Legal and Technical Lessons from a Large-Scale Web-Scraped Machine Learning Dataset

This paper audits a massive web-scraped AI training dataset (DataComp CommonPool, 12.8B samples) and finds substantial amounts of personal and sensitive information despite automated cleaning efforts. The authors show that large-scale internet scraping inevitably captures personally identifiable information (PII), including identity documents, resumes, faces, and sensitive demographic data. They argue that current privacy laws—especially exemptions for “publicly available” data—are poorly equipped to address the risks posed by AI systems trained on indiscriminately scraped web data. The paper concludes that policymakers should regulate dataset creation (“the root”) rather than focusing only on downstream AI models (“the leaves”).

### Using Deep Learning and Google Street View to Estimate the Demographic Makeup of Neighborhoods Across the United States

This paper is an illustration about how publicly available data is being used to group and categorized the average citizen without their explicit knowledge. Using Google Street View images, the group focused on detecting the make, model, year, body type, and estimated price of cars captured in the images. The car data was aggregated and used to train a model to predict median household income, racial composition, education levels, and voting behavior at the neighborhood level. Overall, the paper demonstrates how computational methods and large scale public imagery can imitate or even potentially replace traditional demographic data collection. 

### Private traits and attributes are predictable from digital records of human behavior

This 2013 article is an early exploration of how digital behavior records—specifically Facebook Likes—can be used to infer highly personal traits and attributes with surprising accuracy. Using data from more than 58,000 U.S. Facebook users who also provided demographic information and psychometric test results, the authors built models that predicted characteristics such as age, gender, political views, religion, ethnicity, sexual orientation, personality traits, intelligence, life satisfaction, and some family/background variables.

The study found especially high predictive accuracy for some binary traits. For example, the model distinguished African American and Caucasian American users with 95% accuracy, men and women with 93% accuracy, Democrats and Republicans with 85% accuracy, and homosexual versus heterosexual men with 88% accuracy. For numeric traits, age was predicted especially well, and the trait of Openness was predicted at a level close to the reliability of a standard personality test.

A key point of the paper is that prediction does not require users to explicitly reveal sensitive information. Even ordinary Likes that seem unrelated on their face can, in combination, act as signals for private attributes. The authors emphasize that this has both beneficial and troubling implications: it could improve personalization and recommendation systems, but it also creates serious privacy risks because organizations may infer sensitive traits without a person’s awareness or consent.
